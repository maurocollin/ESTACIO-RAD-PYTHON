```python
# =================================================================
# TEMA 3 - EXERCÍCIO 3: MANIPULAÇÃO DE ARQUIVOS (CONTADOR)
# Objetivo: Contar palavras e linhas de um arquivo .txt,
# verificando se o arquivo existe antes da leitura.
# =================================================================
import os

def analisar_arquivo(nome_arquivo):
    print(f"--- Analisando arquivo: {nome_arquivo} ---")

    try:
        # 1. Verificação de existência (Boas Práticas)
        if not os.path.exists(nome_arquivo):
            raise FileNotFoundError(f"O arquivo '{nome_arquivo}' não foi encontrado no diretório.")

        # 2. Abertura e Processamento
        with open(nome_arquivo, 'r', encoding='utf-8') as arquivo:
            linhas = arquivo.readlines()
            total_linhas = len(linhas)

            # Conta palavras separando por espaços em cada linha
            total_palavras = 0
            for linha in linhas:
                palavras = linha.split()
                total_palavras += len(palavras)

        print(f"Sucesso! O arquivo possui {total_linhas} linhas e {total_palavras} palavras.")

    except FileNotFoundError as e:
        print(f"Erro de Diretório: {e}")
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")

# --- Preparação do Ambiente (Exclusivo para teste no Colab) ---
arquivo_teste = "meu_trabalho.txt"
with open(arquivo_teste, 'w', encoding='utf-8') as f:
    f.write("Olá mundo\nEste é um teste para a disciplina de Python.\nDesenvolvimento Rápido de Aplicações.")

# Execução
if __name__ == "__main__":
    analisar_arquivo(arquivo_teste)
    # Teste de erro:
    # analisar_arquivo("arquivo_que_nao_existe.txt")
```
