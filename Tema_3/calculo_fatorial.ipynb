```python
# =================================================================
# TEMA 3 - EXERCÍCIO 2: FATORIAL COM TRATAMENTO EM LISTAS
# Objetivo: Calcular o fatorial de uma lista tratando números
# negativos e elementos não numéricos.
# =================================================================
import math

def calcular_fatoriais_da_lista(lista_numeros):
    print(f"--- Processando Lista: {lista_numeros} ---")

    for item in lista_numeros:
        try:
            # 1. Verifica se o item é um número (int ou float)
            if not isinstance(item, (int, float)):
                raise ValueError(f"O elemento '{item}' não é um número.")

            # 2. Verifica se o número é negativo
            if item < 0:
                raise ValueError(f"O número {item} é negativo (não existe fatorial de negativos).")

            # 3. Cálculo do fatorial (converte para int pois fatorial é para inteiros)
            num_int = int(item)
            resultado = math.factorial(num_int)
            print(f"Fatorial de {num_int} é: {resultado}")

        except ValueError as ve:
            print(f"Erro de Valor: {ve}")

        except Exception as e:
            print(f"Erro inesperado no item {item}: {e}")

# Execução de teste com uma lista mista (conforme pede o exercício)
if __name__ == "__main__":
    minha_lista = [5, 3, -1, "texto", 4]
    calcular_fatoriais_da_lista(minha_lista)
```
