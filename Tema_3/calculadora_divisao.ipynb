```python
# =================================================================
# TEMA 3 - EXERCÍCIO 1: CALCULADORA COM TRATAMENTO DE EXCEÇÕES
# Objetivo: Realizar divisão tratando erro de zero e tipos de dados.
# =================================================================

def executar_divisao():
    print("--- Sistema de Divisão RAD ---")

    try:
        # Entrada de dados
        num1 = input("Digite o primeiro número (dividendo): ")
        num2 = input("Digite o segundo número (divisor): ")

        # Conversão de tipos (pode gerar ValueError)
        n1 = float(num1)
        n2 = float(num2)

        # Operação matemática (pode gerar ZeroDivisionError)
        resultado = n1 / n2

    except ValueError:
        print("Erro: Entrada inválida. Por favor, digite apenas números.")

    except ZeroDivisionError:
        print("Erro: Não é possível realizar uma divisão por zero.")

    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")

    else:
        # Executado apenas se não houver erros
        print(f"Resultado: {n1} / {n2} = {resultado}")

    finally:
        print("--- Operação Finalizada ---")

# Execução do script
if __name__ == "__main__":
    executar_divisao()
```
