```python
# =================================================================
# TEMA 3 - EXERCÍCIO 4: FORMATADOR E PERSISTÊNCIA DE NOMES
# Objetivo: Ler um nome, formatar (Title Case) e salvar em arquivo,
# tratando erros de validação e diretório.
# =================================================================
import os

def processar_nome_completo():
    print("--- Sistema de Padronização de Nomes ---")

    try:
        # 1. Entrada e Validação Simples
        nome_bruto = input("Digite o nome completo: ").strip()

        if not nome_bruto or len(nome_bruto.split()) < 2:
            raise ValueError("O nome deve conter pelo menos nome e sobrenome.")

        # 2. Transformação (Regra de Negócio: João Silva -> João Silva)
        nome_formatado = nome_bruto.title()

        # 3. Persistência em Arquivo
        nome_arquivo = "nomes_cadastrados.txt"

        # Usamos 'a' (append) para adicionar ao final do arquivo sem apagar o anterior
        with open(nome_arquivo, 'a', encoding='utf-8') as arquivo:
            arquivo.write(nome_formatado + "\n")

        print(f"Sucesso! '{nome_formatado}' foi salvo em {nome_arquivo}.")

    except ValueError as ve:
        print(f"Erro de Validação: {ve}")
    except PermissionError:
        print("Erro: Sem permissão para escrever no diretório.")
    except Exception as e:
        print(f"Erro inesperado: {e}")

# Execução
if __name__ == "__main__":
    processar_nome_completo()
```
