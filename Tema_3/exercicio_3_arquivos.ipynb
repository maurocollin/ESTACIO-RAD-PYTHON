```python
# =================================================================
# TEMA 3 - EXERCÍCIO 1: CALCULADORA COM TRATAMENTO DE EXCEÇÕES
# Objetivo: Realizar divisão tratando erro de zero e tipos de dados.
# =================================================================

def executar_divisao():
    print("--- Sistema de Divisão RAD ---")

    try:
        # Entrada de dados
        num1 = input("Digite o primeiro número (dividendo): ")
        num2 = input("Digite o segundo número (divisor): ")

        # Conversão de tipos (pode gerar ValueError)
        n1 = float(num1)
        n2 = float(num2)

        # Operação matemática (pode gerar ZeroDivisionError)
        resultado = n1 / n2

    except ValueError:
        print("Erro: Entrada inválida. Por favor, digite apenas números.")

    except ZeroDivisionError:
        print("Erro: Não é possível realizar uma divisão por zero.")

    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")

    else:
        # Executado apenas se não houver erros
        print(f"Resultado: {n1} / {n2} = {resultado}")

    finally:
        print("--- Operação Finalizada ---")

# Execução do script
if __name__ == "__main__":
    executar_divisao()
```

```python
# =================================================================
# TEMA 3 - EXERCÍCIO 3: MANIPULAÇÃO DE ARQUIVOS (CONTADOR)
# Objetivo: Contar palavras e linhas de um arquivo .txt,
# verificando se o arquivo existe antes da leitura.
# =================================================================
import os

def analisar_arquivo(nome_arquivo):
    print(f"--- Analisando arquivo: {nome_arquivo} ---")

    try:
        # 1. Verificação de existência (Boas Práticas)
        if not os.path.exists(nome_arquivo):
            raise FileNotFoundError(f"O arquivo '{nome_arquivo}' não foi encontrado no diretório.")

        # 2. Abertura e Processamento
        with open(nome_arquivo, 'r', encoding='utf-8') as arquivo:
            linhas = arquivo.readlines()
            total_linhas = len(linhas)

            # Conta palavras separando por espaços em cada linha
            total_palavras = 0
            for linha in linhas:
                palavras = linha.split()
                total_palavras += len(palavras)

        print(f"Sucesso! O arquivo possui {total_linhas} linhas e {total_palavras} palavras.")

    except FileNotFoundError as e:
        print(f"Erro de Diretório: {e}")
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")

# --- Preparação do Ambiente (Exclusivo para teste no Colab) ---
arquivo_teste = "meu_trabalho.txt"
with open(arquivo_teste, 'w', encoding='utf-8') as f:
    f.write("Olá mundo\nEste é um teste para a disciplina de Python.\nDesenvolvimento Rápido de Aplicações.")

# Execução
if __name__ == "__main__":
    analisar_arquivo(arquivo_teste)
    # Teste de erro:
    # analisar_arquivo("arquivo_que_nao_existe.txt")
```

```python
# =================================================================
# TEMA 3 - EXERCÍCIO 4: FORMATADOR E PERSISTÊNCIA DE NOMES
# Objetivo: Ler um nome, formatar (Title Case) e salvar em arquivo,
# tratando erros de validação e diretório.
# =================================================================
import os

def processar_nome_completo():
    print("--- Sistema de Padronização de Nomes ---")

    try:
        # 1. Entrada e Validação Simples
        nome_bruto = input("Digite o nome completo: ").strip()

        if not nome_bruto or len(nome_bruto.split()) < 2:
            raise ValueError("O nome deve conter pelo menos nome e sobrenome.")

        # 2. Transformação (Regra de Negócio: João Silva -> João Silva)
        nome_formatado = nome_bruto.title()

        # 3. Persistência em Arquivo
        nome_arquivo = "nomes_cadastrados.txt"

        # Usamos 'a' (append) para adicionar ao final do arquivo sem apagar o anterior
        with open(nome_arquivo, 'a', encoding='utf-8') as arquivo:
            arquivo.write(nome_formatado + "\n")

        print(f"Sucesso! '{nome_formatado}' foi salvo em {nome_arquivo}.")

    except ValueError as ve:
        print(f"Erro de Validação: {ve}")
    except PermissionError:
        print("Erro: Sem permissão para escrever no diretório.")
    except Exception as e:
        print(f"Erro inesperado: {e}")

# Execução
if __name__ == "__main__":
    processar_nome_completo()
```
